```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('science')
plt.rcParams['figure.figsize'] = (7,5)
plt.rcParams['font.size'] = 15
plt.rcParams['text.usetex'] = True
```

```python
t = 0.1
eps = 1.0
a = 1.0
```

```python
def E(k, eps, t, a):
    return eps - 2*t*np.cos(k*a)

k = np.linspace(-np.pi/a, np.pi/a, 100)

plt.plot(k, E(k, eps, t, a))
plt.xlabel('$k$')
plt.ylabel('$E(k)$')
plt.title('$\epsilon = 1.0, a = 1.0, t = 0.1.$')
plt.grid()
plt.savefig('fig/E.png')
plt.show()
```

```python
def d_infinite(E, t, eps):
    return 1 / (np.pi * np.sqrt(4*t**2 - (E - eps)**2))

E = np.linspace(eps - abs(t), eps + abs(t), 100)

plt.plot(E, d_infinite(E, t, eps))
plt.xlabel('$E$')
plt.ylabel('$d(E)$')
plt.title('$\epsilon = 1.0, t = 0.1$')
plt.grid()
plt.savefig('fig/d-infinite.png')
plt.show()
```

```python
def d_finite(E, t, eps):
    return np.sqrt(4*t**2 - (E - eps)**2) / (2*np.pi*t**2)

E = np.linspace(eps - abs(t), eps + abs(t), 100)

plt.plot(E, d_finite(E, t, eps))
plt.xlabel('$E$')
plt.ylabel('$d_1(E)$')
plt.title('$\epsilon = 1.0, t = 0.1$')
plt.grid()
plt.savefig('fig/d-finite.png')
plt.show()
```
