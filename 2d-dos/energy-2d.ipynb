```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1.0
alpha=0.0
beta=-1.0

kx = np.linspace(-np.pi/a, np.pi/a, 51)
ky = np.linspace(-np.pi/a, np.pi/a, 51)

kxv, kyv = np.meshgrid(kx, ky)

def E(kxv, kyv, a=1.0, alpha=0.0, beta=-1.0):
    return alpha + 2*beta*(np.cos(kxv*a) + np.cos(kyv*a))
```

```python
%matplotlib notebook
ax = plt.figure(figsize=(8,8)).add_subplot(projection='3d')
ax.scatter(kxv, kyv, E(kxv, kyv))
ax.set_xlabel('$k_x$')
ax.set_ylabel('$k_y$')
ax.set_zlabel('$E(k)$')
plt.show()
```

```python
%matplotlib inline

s_1 =(np.linspace(0, np.pi/a), 0)
s_2 =(np.linspace(np.pi/a, 0), np.linspace(0, np.pi/a))
s_3 =(0, np.linspace(np.pi/a, 0))

E_cut = np.hstack((E(s_1[0], s_1[1]), E(s_2[0], s_2[1]), E(s_3[0], s_3[1])))

plt.plot(E_cut)
plt.ylabel('E')
plt.xticks([0, 50, 100, 150], ['$\Gamma$', '$X$', '$M$', '$\Gamma$'])
plt.grid()
plt.show()
```
