```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def E(kx, ky, eps=0, beta=2.8, a=1.0):
    return beta*np.sqrt(1 + 4*np.cos(kx*a/2)**2 + 4*np.cos(kx*a/2)*np.cos(ky*a*np.sqrt(3)/2))
```

```python
a = 1.0
kx = np.linspace(-2*np.pi/a, 2*np.pi/a, 101)
ky = np.linspace(-2*np.pi/a, 2*np.pi/a, 101)

kxv, kyv = np.meshgrid(kx, ky)
```

```python
%matplotlib notebook

ax = plt.figure(figsize=(8,8)).add_subplot(projection='3d')
ax.plot_surface(kxv, kyv, E(kxv, kyv))
ax.plot_surface(kxv, kyv, -E(kxv, kyv))
ax.set_xlabel('$k_x$')
ax.set_ylabel('$k_y$')
ax.set_zlabel('$E(k)$')
plt.show()
```

```python
%matplotlib inline

s_1 = (0, np.linspace(0, 2*np.pi*np.sqrt(3)/3*a))
s_2 = (np.linspace(0, np.pi/a), 2*np.pi*np.sqrt(3)/3*a)
s_3 = (np.linspace(np.pi/a, 0), np.linspace(2*np.pi*np.sqrt(3)/3*a, 0))

E_cut = np.hstack((E(s_1[0], s_1[1]), E(s_2[0], s_2[1]), E(s_3[0], s_3[1])))

plt.plot(E_cut)
plt.plot(-E_cut)
plt.ylabel('E')
plt.xticks([0, 50, 100, 150], ['$\Gamma$', '$M$', '$K$', '$\Gamma$'])
plt.grid()
plt.show()
```
