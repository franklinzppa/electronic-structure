```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def E(kxv, kyv, beta=-1, a=1):
    return beta*np.sqrt(1 + 4*np.cos(kxv*a/2)**2 + 4*np.cos(kxv*a/2)*np.cos(kyv*a*np.sqrt(3)/2))
```

```python
a = 1.0
kx = np.linspace(-3*np.pi/a, 3*np.pi/a, 101)
ky = np.linspace(-3*np.pi/a, 3*np.pi/a, 101)

kxv, kyv = np.meshgrid(kx, ky)
```

```python
%matplotlib notebook
ax = plt.figure(figsize=(8,8)).add_subplot(projection='3d')
ax.scatter(kxv, kyv, E(kxv, kyv))
ax.scatter(kxv, kyv, -E(kxv, kyv))
ax.set_xlabel('$k_x$')
ax.set_ylabel('$k_y$')
ax.set_zlabel('$E(k)$')
plt.show()
```

```python
%matplotlib inline

s_1 =(0, np.linspace(0, np.pi/a))
s_2 =(np.linspace(0, np.pi*np.sqrt(3)*a), np.pi/a)
s_3 =(np.linspace(np.pi*np.sqrt(3)*a, 0), np.linspace(np.pi/a, 0))

E_cut = np.hstack((E(s_1[0], s_1[1]), E(s_2[0], s_2[1]), E(s_3[0], s_3[1])))

plt.plot(E_cut)
plt.plot(-E_cut)
plt.ylabel('E')
plt.xticks([0, 50, 100, 150], ['$\Gamma$', '$M$', '$K$', '$\Gamma$'])
plt.grid()
plt.show()
```
